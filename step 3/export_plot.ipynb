```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("/content/data.xlsx")

# df[df['RegionType'].str.contains('SoyBean', case=False)]

# Indonesia-oil & palm fruit, SE. Asia-oil & palm frui, SE. Asia-rice, Brazil-soybeans
```

```python
df_melt = df.melt(id_vars=['RegionType', 'type'], var_name='time', value_name='value')
df_melt
```

```python
df_melt[(df_melt['RegionType'] == "Brazil-soybeans")]
```

```python
# Define custom color mapping
color_mapping = {
    "Indonesia-oil & palm fruit": sns.color_palette("muted")[0],
    "SE. Asia-oil & palm fruit": sns.color_palette("muted")[1],
    "SE. Asia-rice": sns.color_palette("muted")[2],
    "Brazil-soybeans": sns.color_palette("muted")[3],
}
palette = {}
for k in df_melt['RegionType'].unique():
    if k in color_mapping:
        palette[k] = color_mapping[k]
    else:
        palette[k] = 'lightgrey'

plt.figure(figsize=(10,6))
sns.lineplot(
    data=df_melt[(df_melt['RegionType'].isin(["Indonesia-oil & palm fruit", "SE. Asia-oil & palm fruit", "SE. Asia-rice", "Brazil-soybeans"]))],
    x="time", y="value", hue="RegionType",
    hue_order=["Indonesia-oil & palm fruit", "SE. Asia-oil & palm fruit", "SE. Asia-rice", "Brazil-soybeans"],
    palette={k: palette[k] for k in ["Indonesia-oil & palm fruit", "SE. Asia-oil & palm fruit", "SE. Asia-rice", "Brazil-soybeans"]},
    estimator=None, dashes=False, markers=True
)

sns.lineplot(
    data=df_melt[~(df_melt['RegionType'].isin(["Indonesia-oil & palm fruit", "SE. Asia-oil & palm fruit", "SE. Asia-rice", "Brazil-soybeans"]))],
    x="value", y="time", hue="RegionType",
    palette={k: 'lightgrey' for k in df_melt['RegionType'].unique() if k not in ["Indonesia-oil & palm fruit", "SE. Asia-oil & palm fruit", "SE. Asia-rice", "Brazil-soybeans"]},
    estimator=None, dashes=False, markers=True, legend=False
)

plt.legend(loc='upper left')
plt.show()
```
