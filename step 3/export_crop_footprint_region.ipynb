```python
import numpy as np
import pandas as pd
```

```python
am1crop_emisioncrops_gfrac = pd.read_excel("D:/kerja/asisten riset/code v/step 3/GbGlct/AM1_prop_emission_crops_GFRAC_afterELKE_run3a.xlsx")
fao_area_harvested = pd.read_excel("D:/kerja/asisten riset/vol/milkunC/achaidir/FAO/faostat_area_harvested_IMAGE.xlsx")
ccode_iso = pd.read_excel("D:/kerja/asisten riset/vol/milkunC/achaidir/LUH2 2022/ISO-3166-Country-Code_Final.xlsx")

am1crop_emisioncrops_gfrac['country_name'] = am1crop_emisioncrops_gfrac['country_name'].replace("United Kingdom of Great Britain and Northern Irela", "United Kingdom of Great Britain and Northern Ireland")
am1crop_emisioncrops_gfrac = am1crop_emisioncrops_gfrac.rename(columns=lambda x: x.capitalize())
am1crop_emisioncrops_gfrac = am1crop_emisioncrops_gfrac.rename(columns={'Grass': 'grass'})
```

```python
fao_area_harvested_columns = fao_area_harvested[['IMAGE Region Name', 'Year', 'IMAGE Classification', 'Value']]
fao_area_harvested_columns.head()
```

```python
am1crop_emisioncrops_gfrac.rename(columns={"Country_name": "Country"}, inplace=True)
```

```python
am1crop_emisioncrops_gfrac_merge = pd.merge(left=am1crop_emisioncrops_gfrac, right=ccode_iso, how="left", left_on="Country", right_on="ISO Country")
am1crop_emisioncrops_gfrac_merge.columns
```

```python
am1crop_emisioncrops_gfrac_merge_groubpy = am1crop_emisioncrops_gfrac_merge.groupby(['IMAGE Region Name', 'Time']).sum()
am1crop_emisioncrops_gfrac_merge_groubpy = am1crop_emisioncrops_gfrac_merge_groubpy.drop(columns=["Country", "IMAGE Region Code",  "iso_3166-2", 'ISO Country', 'alpha-2', 'alpha-3', 'country-code', "Sub Region", "ISO Region"])
am1crop_emisioncrops_gfrac_merge_groubpy =  am1crop_emisioncrops_gfrac_merge_groubpy.reset_index()
am1crop_emisioncrops_gfrac_merge_groubpy['Time'] = am1crop_emisioncrops_gfrac_merge_groubpy['Time'].astype('int')
am1crop_emisioncrops_gfrac_merge_groubpy.head()
```

```python
fao_area_harvested_merge_groubpy = fao_area_harvested.groupby(['IMAGE Region Name', 'Year', 'IMAGE Classification'])['Value'].sum()
# fao_area_harvested_merge_groubpy = fao_area_harvested_merge_groubpy.drop(columns=["ISO Country", "Sub Region"])
fao_area_harvested_merge_groubpy =  fao_area_harvested_merge_groubpy.reset_index()
fao_area_harvested_merge_groubpy['Year'] = fao_area_harvested_merge_groubpy['Year'].astype('int')
fao_area_harvested_merge_groubpy.head()
```

```python
fao_area_harvested_copy = fao_area_harvested_merge_groubpy.copy()

for negara in np.unique(fao_area_harvested_merge_groubpy['IMAGE Region Name'].to_list()):
    for year in range(1975, 2025, 5):
        for ngfbfc in np.unique(fao_area_harvested_merge_groubpy['IMAGE Classification'].to_list()):
            try:
                am1crop_emisioncrops_gfrac_values = am1crop_emisioncrops_gfrac_merge_groubpy[(am1crop_emisioncrops_gfrac_merge_groubpy['IMAGE Region Name'] == negara) & (am1crop_emisioncrops_gfrac_merge_groubpy['Time'] == year)][ngfbfc].values
                fao_area_harvested_values = fao_area_harvested_merge_groubpy[(fao_area_harvested_merge_groubpy['Year'] == year) & (fao_area_harvested_merge_groubpy['IMAGE Region Name'] == negara) & (fao_area_harvested_merge_groubpy['IMAGE Classification'] == ngfbfc)]['Value'].values
                fao_area_harvested_copy.loc[fao_area_harvested_merge_groubpy[(fao_area_harvested_merge_groubpy['IMAGE Region Name'] == negara) & (fao_area_harvested_merge_groubpy['Year'] == year) & (fao_area_harvested_merge_groubpy['IMAGE Classification'] == ngfbfc)].index, 'Value'] = am1crop_emisioncrops_gfrac_values/fao_area_harvested_values
            except:
                print(negara)
```

```python
fao_area_harvested_copy = fao_area_harvested_copy[fao_area_harvested_copy['Year'] > 1970]
fao_area_harvested_copy = fao_area_harvested_copy[["IMAGE Region Name", "Year", "IMAGE Classification", "Value"]]
fao_area_harvested_copy
```

```python
fao_area_harvested_copy.to_excel("D:/kerja/asisten riset/code v/step 3/GbGlct/AM1_emission_crops_footprint_region.xlsx", index=False)
```
