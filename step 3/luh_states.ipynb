```python
import numpy as np
import xarray
import pandas as pd
```

```python
# ### States
path_LUH2 = 'D:/kerja/asisten riset/vol/milkunC/achaidir/LUH2 2022/states.nc'
luh2_states = xarray.open_dataset(path_LUH2, engine="netcdf4", decode_times=False)
luh2_states_worldwide = luh2_states.isel(time=slice(1120, 1173))
luh2_states_worldwide['time'] = pd.date_range(start="1970-01-01", end="2022-01-01", freq='YS')
luh2_states_worldwide
```

```python
new_expand_c3ann = np.expand_dims(luh2_states_worldwide['c3ann'].values, axis=-1)
new_expand_c4ann = np.expand_dims(luh2_states_worldwide['c4ann'].values, axis=-1)
new_expand_c3per = np.expand_dims(luh2_states_worldwide['c3per'].values, axis=-1)
new_expand_c4per = np.expand_dims(luh2_states_worldwide['c4per'].values, axis=-1)
new_expand_c3nfx = np.expand_dims(luh2_states_worldwide['c3nfx'].values, axis=-1)

new_luh_states = np.concatenate([new_expand_c3ann, new_expand_c4ann, new_expand_c3per,
                                    new_expand_c4per, new_expand_c3nfx], axis=-1)
```

```python
landuse_class_crops = ['c3ann', 'c4ann', 'c3per', 'c4per', 'c3nfx']

new_luh2states = xarray.Dataset({
        "luh2_states":(["time", "lat", "lon", "cat"], new_luh_states)
    },
    coords={
        "lon":luh2_states_worldwide.coords["lon"].to_numpy(),
        "lat":luh2_states_worldwide.coords["lat"].to_numpy(),
        "time": pd.date_range(start="1970-01-01", end="2022-01-01", freq='YS'),
        "cat": landuse_class_crops
    })
# new_luh2states
new_luh2states.to_netcdf("D:/kerja/asisten riset/vol/milkunC/achaidir/LUH2 2022/states_new.nc", mode='w', format="NETCDF4")
```
