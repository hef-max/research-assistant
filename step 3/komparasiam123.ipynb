```python
import pandas as pd
import numpy as np
```

```python
am1 = pd.read_excel("D:/kerja/asisten riset/vol/milkunC/achaidir/nco2/emisi_total_per_crops_AM1_region_v2.xlsx", sheet_name="Sheet1")
am2 = pd.read_excel("D:/kerja/asisten riset/vol/milkunC/achaidir/nco2/emisi_total_per_crops_AM2_region_v2.xlsx", sheet_name="Sheet1")
am3 = pd.read_excel("D:/kerja/asisten riset/vol/milkunC/achaidir/nco2/emisi_total_per_crops_AM3_v2.xlsx", sheet_name="Sheet1")
am4 = pd.read_excel("D:/kerja/asisten riset/vol/milkunC/achaidir/nco2/emisi_total_per_crops_AM4.xlsx", sheet_name="Sheet1")
```

```python
am1_melt = am1.melt(id_vars=['IMAGE Region Name', 'type'], var_name='time', value_name='value')
am1_melt.loc[:, 'type'] = am1_melt['type'].str.replace("other non-food & luxury & spices", "non food, luxury, spices")
df_am1_groupby = am1_melt.groupby(['IMAGE Region Name', 'type']).sum()
df_am1_groupby = df_am1_groupby.drop(columns=['time'])
df_am1_index = df_am1_groupby.reset_index()

am1_pivot = pd.pivot_table(df_am1_index, index=['IMAGE Region Name'], columns=['type'], fill_value=0)
am1_stack = am1_pivot.stack(level=0, future_stack=True )
am1_index = am1_stack.reset_index()
am1_index = am1_index.drop(columns=['level_1'])
am1_index
```

```python
am2.loc[:, 'type'] = am2['type'].str.replace("other non-food & luxury & spices", "non food, luxury, spices")
am2_pivot = pd.pivot_table(am2, index=['IMAGE Region Name'], columns=['type'], fill_value=0)
am2_stack = am2_pivot.stack(level=0, future_stack=True )
am2_index = am2_stack.reset_index()
am2_index = am2_index.drop(columns=['level_1'])
am2_index
```

```python
am3 = am3[(~am3['IMAGE_Region_Name'].isna()) & (am3['IMAGE_Region_Name'] != 'ocean')]
am3.loc[:, 'type'] = am3['type'].str.replace("other non-food & luxury & spices", "non food, luxury, spices")
am3_pivot = pd.pivot_table(am3, index=['IMAGE_Region_Name'], columns=['type'], fill_value=0)
am3_stack = am3_pivot.stack(level=0, future_stack=True )
am3_index = am3_stack.reset_index()
am3_index = am3_index.drop(columns=['level_1'])
am3_index
```

```python
am4 = am4[(~am4['IMAGE Region Name'].isna()) & (am4['IMAGE Region Name'] != 'ocean') & (am4['IMAGE Region Name'] != 'N|A')]
am4.loc[:, 'type'] = am4['type'].str.replace("other non-food & luxury & spices", "non food, luxury, spices")
am4_pivot = pd.pivot_table(am4, index=['IMAGE Region Name'], columns=['type'], fill_value=0)
am4_stack = am4_pivot.stack(level=0, future_stack=True )
am4_index = am4_stack.reset_index()
am4_index = am4_index.drop(columns=['level_1'])
am4_index
```

```python
column = am1_index.columns[1:].to_list()
komparasi = pd.DataFrame(columns=column, index=["AM1 (%)", "AM2 (%)", "AM3 (%)", "AM4 (%)"])
komparasi.index.name = "Crop Category"
komparasi = komparasi.reset_index()

with pd.ExcelWriter("D:/kerja/asisten riset/vol/milkunC/achaidir/nco2/KomparasiAM1234_versi2.xlsx", engine='xlsxwriter') as writer:
    for region in am3_index['IMAGE_Region_Name']:
        komparasi.iloc[0, 1:] = am1_index[am1_index['IMAGE Region Name'] == region].to_numpy()[0][1:]/sum(am1_index[am1_index['IMAGE Region Name'] == region].to_numpy()[0][1:]) * 100
        komparasi.iloc[1, 1:] = am2_index[am2_index['IMAGE Region Name'] == region].to_numpy()[0][1:]/sum(am2_index[am2_index['IMAGE Region Name'] == region].to_numpy()[0][1:]) * 100
        komparasi.iloc[2, 1:] = am3_index[am3_index['IMAGE_Region_Name'] == region].to_numpy()[0][1:]/sum(am3_index[am3_index['IMAGE_Region_Name'] == region].to_numpy()[0][1:]) * 100
        komparasi.iloc[3, 1:] = am4_index[am4_index['IMAGE Region Name'] == region].to_numpy()[0][1:]/sum(am4_index[am4_index['IMAGE Region Name'] == region].to_numpy()[0][1:]) * 100
        komparasi.to_excel(writer, sheet_name=f"{region}", index=False)
```
