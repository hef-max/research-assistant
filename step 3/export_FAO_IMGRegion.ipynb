```python
import numpy as np
import pandas as pd
```

```python
# ccode_iso = pd.read_excel("D:/kerja/asisten riset/vol/milkunC/achaidir/LUH2 2022/ISO-3166-Country-Code.xlsx")
df_IMAGE = pd.read_excel("D:/kerja/asisten riset/code v/step 5/IMAGE Region.xlsx", sheet_name="Sheet1")
ccode_rev = pd.read_excel("D:/kerja/asisten riset/vol/milkunC/achaidir/LUH2 2022/ISO-3166-Country-Code_REV2.xlsx", sheet_name="country")
```

```python
ccode_rev = ccode_rev.loc[0:247]
missing_rows = ccode_rev.isna().any(axis=1)
ccode_rev.loc[missing_rows,'alpha-2'] = "NA"
ccode_rev.isna().sum()
```

```python
missing_rows = df_IMAGE.isna().any(axis=1)
```

```python
df_IMAGE[missing_rows]
```

```python
df = df_IMAGE.copy()
df = df.fillna("N/A")
df.isna().sum()
```

```python
len(df['ISO Code'].unique())
```

```python
len(ccode_rev['alpha-3'].unique())
```

```python
ccode_merge = pd.merge(left=ccode_rev, right=df, left_on="alpha-3", right_on="ISO Code")
ccode_merge.isna().sum()
```

```python
ccode_merge.columns
```

```python
ccode_merge_columns = ccode_merge[['ISO Country', 'alpha-2', 'alpha-3', 'country-code', 'iso_3166-2',
       'Sub Region', 'ISO Region', 'Native Region Code', 'Native Region Name']]
ccode_merge_columns.head()
```

```python
ccode_merge_copy = ccode_merge_columns.copy()
ccode_merge_copy['country-code'] = ccode_merge_copy['country-code'].astype('int64')
ccode_merge_copy.head()
```

```python
ccode_merge_copy_columns = ccode_merge_copy.rename(columns = {"Native Region Code": "IMAGE Region Code", "Native Region Name": "IMAGE Region Name"})
ccode_merge_copy_columns.head()
```

```python
ccode_merge_copy_columns = ccode_merge_copy_columns.apply(lambda x: x.replace("N/A", "N|A"))
ccode_merge_copy_columns[missing_rows]
```

```python
ccode_merge_copy_columns.to_excel("D:/kerja/asisten riset/vol/milkunC/achaidir/LUH2 2022/ISO-3166-Country-Code_Final.xlsx", index=False)
```
