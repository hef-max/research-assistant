```python
df_new_copy = df_new_1k_rows.copy()

# step 1
def fill_nearest_step1(row, df):
    if pd.isna(row['N_rate_kg_ha']):
        matching_rows = df[
            (df['IMAGE Region Name'] == row['IMAGE Region Name']) &
            (df['Year'] == row['Year']) &
            (df['IMAGE Classification'] == row['IMAGE Classification']) &
            (~df['N_rate_kg_ha'].isna())
        ]
        
        if not matching_rows.empty:
            return matching_rows['N_rate_kg_ha'].mean()
        
    return row['N_rate_kg_ha']

df_new_copy['N_rate_kg_ha'] = df_new_copy.apply(fill_nearest_step1, axis=1, df=df_new_copy)
df_new_copy
```

```python
df_new_copy2 = df_new_copy.copy()

# step 2
def fill_nearest_step2(row, df):
    if pd.isna(row['N_rate_kg_ha']):
        matching_rows = df[
            (df['country'] == row['country']) &
            (df['IMAGE Region Name'] == row['IMAGE Region Name']) &
            (df['FAO Crops'] == row['FAO Crops']) &
            (df['IMAGE Classification'] == row['IMAGE Classification'])
        ]
        
        if not matching_rows.empty:
            # Melakukan forward fill pada 'N_rate_kg_ha' dan mengambil nilai pertama yang diisi
            filled_value = matching_rows['N_rate_kg_ha'].ffill().iloc[0]
            return filled_value
        
    return row['N_rate_kg_ha']

df_new_copy2['N_rate_kg_ha'] = df_new_copy2.apply(fill_nearest_step2, axis=1, df=df_new_copy2)
df_new_copy2
```

```python
import xarray as xr
import numpy as np

# Buat dictionary kode negara ke nama negara
ccode_iso = list(country_code['country-code'])
cname_iso = list(country_code['ISO Country'])

ccode_dict = {ccode: cname for ccode, cname in zip(ccode_iso, cname_iso)}

# Ubah array kode negara ke nama negara langsung dengan vectorized mapping
ccode_worldwide_int = luh_static['ccode'].to_numpy().astype('int64')
ccode_names = np.vectorize(ccode_dict.get)(ccode_worldwide_int, "ocean")  # Ganti yang tidak ada dengan "ocean"

# Buat array kosong untuk NetCDF
fubc_netcdf = np.zeros((11, 14, 2160, 4320), dtype='float32')

# Loop hanya pada waktu dan NGFBFC saja, untuk spatial grid gunakan indexing langsung
for x in range(0, 11):
    for y, ngfbfc in enumerate(fubc_ngfbfc):
        # Ambil data `N_rate_kg_ha` untuk waktu `x` dan kategori `ngfbfc`
        data_slice = df_xarray['N_rate_kg_ha'].isel(time=x).sel(NGFBFC=ngfbfc)
        
        # Gunakan numpy indexing dengan matching nama negara
        mask = np.isin(ccode_names, data_slice['country'].values)
        fubc_netcdf[x, y][mask] = data_slice.sel(country=ccode_names[mask]).values
```

```python
# Buat kembali dataset xarray dari fubc_netcdf jika dibutuhkan
fubc_xarray = xr.DataArray(
    fubc_netcdf, 
    dims=["time", "NGFBFC", "latitude", "longitude"],
    coords={
        "time": pd.date_range(start='1970-01-01', end='2020-01-01', freq='5YS'),
        "NGFBFC": fubc_ngfbfc,
        "latitude": luh_static.coords["latitude"].to_numpy(),
        "longitude": luh_static.coords["longitude"].to_numpy(),
    }
)
fubc_xarray
```
