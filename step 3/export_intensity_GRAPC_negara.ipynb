```python
import numpy as np
import pandas as pd
```

```python
am1crop_emisioncrops_gfrac = pd.read_excel("D:/kerja/asisten riset/code v/step 3/GbGlct/AM1_prop_emission_crops_GFRAC_afterELKE_run3a.xlsx")
faostat_yield = pd.read_csv("D:/kerja/asisten riset/vol/milkunC/achaidir/FAO/faostat_yield.csv")

am1crop_emisioncrops_gfrac['country_name'] = am1crop_emisioncrops_gfrac['country_name'].replace("United Kingdom of Great Britain and Northern Irela", "United Kingdom of Great Britain and Northern Ireland")
am1crop_emisioncrops_gfrac = am1crop_emisioncrops_gfrac.rename(columns=lambda x: x.capitalize())
am1crop_emisioncrops_gfrac = am1crop_emisioncrops_gfrac.rename(columns={'Grass': 'grass'})
```

```python
faostat_yield_copy = faostat_yield.copy()
faostat_yield_copy['New Value'] = 0.0
faostat_yield_copy['New Value'].astype('float32')

for negara in np.unique(faostat_yield['Area'].to_list()):
    for year in range(1975, 2025, 5):
        for ngfbfc in np.unique(faostat_yield['IMAGE Classification'].to_list()):
            try:
                am1crop_emisioncrops_gfrac_values = am1crop_emisioncrops_gfrac[(am1crop_emisioncrops_gfrac['Country_name'] == negara) & (am1crop_emisioncrops_gfrac['Time'] == year)][ngfbfc].values
                faostat_yield_values = faostat_yield[(faostat_yield['Year'] == year) & (faostat_yield['Area'] == negara) & (faostat_yield['IMAGE Classification'] == ngfbfc)]['True Value'].values
                faostat_yield_copy.loc[faostat_yield[(faostat_yield['Area'] == negara) & (faostat_yield['Year'] == year) & (faostat_yield['IMAGE Classification'] == ngfbfc)].index, 'New Value'] = am1crop_emisioncrops_gfrac_values/faostat_yield_values
            except:
                print(negara)
```

```python
faostat_yield_copy = faostat_yield_copy[faostat_yield_copy['Year'] > 1970]
faostat_yield_copy = faostat_yield_copy[["Area", "Element", "Year", "IMAGE Classification", "New Value"]]
faostat_yield_copy
```
