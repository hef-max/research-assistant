```python
import pandas as pd
import numpy as np
```

```python
emisitotal = pd.read_excel("D:/kerja/asisten riset/code v/step 3/GbGlct/dataset/AM3/Emisitotal_RegionType_AM3.xlsx")
areaharvd = pd.read_csv("D:/kerja/asisten riset/vol/milkunC/achaidir/IMAGE PBL/SSP2/GFRAC Area HarvestedRegionType_fulll.csv")
```

```python
emisitotal[2020] = emisitotal[2020] * 1E9
emisitotal
```

```python
areaharvd_2020 = areaharvd[areaharvd['time'] == 2020]
areaharvd_2020.head()
```

```python
# 1 b c
df_copy = areaharvd_2020.copy()
for region, time, ngfbfc in zip(df_copy["IMAGE Region Name"].to_list(), df_copy["time"].to_list(), df_copy["NGFBFC"].to_list()):
    try:
        data1 = areaharvd_2020[(areaharvd_2020['IMAGE Region Name'] == region) & (areaharvd_2020['time'] == time) & (areaharvd_2020['NGFBFC'] == ngfbfc)]['value'].to_numpy()
        data2 = emisitotal[(emisitotal['region_name'] == region) & (emisitotal['type'] == ngfbfc)][time].to_numpy()
        if data1.size == 0:
            data1 = [0]
        if data2.size == 0: 
            data2 = [0]
        mask = df_copy[(df_copy['IMAGE Region Name'] == region) & (df_copy['time'] == time) & (df_copy['NGFBFC'] == ngfbfc)].index
        df_copy.loc[mask, 'Carbon Footprint'] =  data2[0] / data1[0]
    except KeyError or  ZeroDivisionError:
        print(f"Error: {KeyError} for region {region}")

df_copy = df_copy.drop(columns=['value'])
df_copy.head()
```

```python
df_copy.to_csv("D:/kerja/asisten riset/vol/milkunC/achaidir/FAO/carbon_footprint_am3.csv", index=False)
```
