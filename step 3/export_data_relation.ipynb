```python
columns = df.columns[2:].to_list()
df_pivot = df.pivot_table(columns='time', values=columns, aggfunc="sum")
df_pivot.head()
```

```python
df_pivot_index = df_pivot.reset_index()
df_pivot_index
```

```python
df_pivot_index = df_pivot_index.rename(columns={"index":"type"})
df_pivot_index.head()
```

```python
columns_time = df_pivot_index.columns[2:].to_list()
df_melt = df_pivot_index.melt(id_vars='type', var_name="time", value_name='value')
df_melt.head()
```
