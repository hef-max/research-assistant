```python
import numpy as np
import pandas as pd
```

```python
am1crop_emisioncrops_gfrac = pd.read_excel("D:/kerja/asisten riset/code v/step 3/GbGlct/AM1_prop_emission_crops_GFRAC_afterELKE_run3a.xlsx")
faostat_yield = pd.read_csv("D:/kerja/asisten riset/vol/milkunC/achaidir/FAO/faostat_yield.csv")
ccode_iso = pd.read_excel("D:/kerja/asisten riset/vol/milkunC/achaidir/LUH2 2022/ISO-3166-Country-Code_REV.xlsx", sheet_name="Sheet3")

am1crop_emisioncrops_gfrac['country_name'] = am1crop_emisioncrops_gfrac['country_name'].replace("United Kingdom of Great Britain and Northern Irela", "United Kingdom of Great Britain and Northern Ireland")
am1crop_emisioncrops_gfrac = am1crop_emisioncrops_gfrac.rename(columns=lambda x: x.capitalize())
am1crop_emisioncrops_gfrac = am1crop_emisioncrops_gfrac.rename(columns={'Grass': 'grass'})
```

```python
am1crop_emisioncrops_gfrac.rename(columns={"Country_name": "Country"}, inplace=True)
faostat_yield.rename(columns={"Area": "Country"}, inplace=True)
```

```python
am1crop_emisioncrops_gfrac_merge = pd.merge(left=ccode_iso, right=am1crop_emisioncrops_gfrac, how="left", left_on="Country", right_on="Country")
faostat_yield_merge = pd.merge(left=ccode_iso, right=faostat_yield, how="left", left_on="Country", right_on="Country")
```

```python
am1crop_emisioncrops_gfrac_merge_groubpy = am1crop_emisioncrops_gfrac_merge.groupby(['Region', 'Time']).sum()
am1crop_emisioncrops_gfrac_merge_groubpy = am1crop_emisioncrops_gfrac_merge_groubpy.drop(columns=["Country", "SubRegion"])
am1crop_emisioncrops_gfrac_merge_groubpy =  am1crop_emisioncrops_gfrac_merge_groubpy.reset_index()
am1crop_emisioncrops_gfrac_merge_groubpy['Time'] = am1crop_emisioncrops_gfrac_merge_groubpy['Time'].astype('int')
am1crop_emisioncrops_gfrac_merge_groubpy
```

```python
faostat_yield_merge = faostat_yield_merge[faostat_yield_merge['Year'] > 1970]
```

```python
faostat_yield_merge_groubpy = faostat_yield_merge.groupby(['Region', 'Year', 'IMAGE Classification']).sum()
faostat_yield_merge_groubpy = faostat_yield_merge_groubpy.drop(columns=["Country", "SubRegion", "Element"])
faostat_yield_merge_groubpy =  faostat_yield_merge_groubpy.reset_index()
faostat_yield_merge_groubpy['Year'] = faostat_yield_merge_groubpy['Year'].astype('int')
```

```python
faostat_yield_copy = faostat_yield_merge_groubpy.copy()
faostat_yield_copy['New Value'] = 0.0
faostat_yield_copy['New Value'].astype('float32')

for negara in np.unique(faostat_yield_merge_groubpy['Region'].to_list()):
    for year in range(1975, 2025, 5):
        for ngfbfc in np.unique(faostat_yield_merge_groubpy['IMAGE Classification'].to_list()):
            try:
                am1crop_emisioncrops_gfrac_values = am1crop_emisioncrops_gfrac_merge_groubpy[(am1crop_emisioncrops_gfrac_merge_groubpy['Region'] == negara) & (am1crop_emisioncrops_gfrac_merge_groubpy['Time'] == year)][ngfbfc].values
                faostat_yield_merge_groubpy_values = faostat_yield_merge_groubpy[(faostat_yield_merge_groubpy['Year'] == year) & (faostat_yield_merge_groubpy['Region'] == negara) & (faostat_yield_merge_groubpy['IMAGE Classification'] == ngfbfc)]['True Value'].values
                faostat_yield_copy.loc[faostat_yield_merge_groubpy[(faostat_yield_merge_groubpy['Region'] == negara) & (faostat_yield_merge_groubpy['Year'] == year) & (faostat_yield_merge_groubpy['IMAGE Classification'] == ngfbfc)].index, 'New Value'] = am1crop_emisioncrops_gfrac_values/faostat_yield_merge_groubpy_values
            except:
                print(negara)
```

```python
faostat_yield_copy = faostat_yield_copy[faostat_yield_copy['Year'] > 1970]
faostat_yield_copy = faostat_yield_copy[["Region", "Year", "IMAGE Classification", "New Value"]]
faostat_yield_copy
```

```python
faostat_yield_copy.to_excel("D:/kerja/asisten riset/code v/step 3/GbGlct/AM1_emission_intensity_region.xlsx", index=False)
```
