```python
import pandas as pd
import numpy as np
```

```python
ccode_iso = pd.read_excel("D:/kerja/asisten riset/vol/milkunC/achaidir/LUH2 2022/ISO-3166-Country-Code_Final.xlsx")
area_harvested = pd.read_csv("D:/kerja/asisten riset/vol/milkunC/achaidir/FAO/faostat_area_harvested.csv")
yield_absolute = pd.read_csv("D:/kerja/asisten riset/vol/milkunC/achaidir/FAO/faostat_yield_absolute.csv")
```

```python
ccode_iso.head()
```

### area harvested

```python
area_harvested.head()
```

```python
area_harvested_merge = pd.merge(left=area_harvested, right=ccode_iso, left_on="Area", right_on="ISO Country")
area_harvested_merge.head()
```

```python
area_harvested_merge_columns = area_harvested_merge[['ISO Country', 'alpha-2', 'alpha-3', 'country-code', 'iso_3166-2', 'Sub Region',
       'ISO Region', 'IMAGE Region Code', 'IMAGE Region Name','Year', 'IMAGE Classification', 'Value']]
area_harvested_merge_columns.head()
```

### yield absolute

```python
yield_absolute.head()
```

```python
yield_absolute_merge = pd.merge(left=yield_absolute, right=ccode_iso, left_on="Area", right_on="ISO Country")
yield_absolute_merge.head()
```

```python
yield_absolute_merge_columns = yield_absolute_merge[['ISO Country', 'alpha-2', 'alpha-3', 'country-code', 'iso_3166-2', 'Sub Region',
       'ISO Region', 'IMAGE Region Code', 'IMAGE Region Name','Year', 'IMAGE Classification', 'Value']]
yield_absolute_merge_columns.head()
```

```python
area_harvested_merge_columns.to_excel("D:/kerja/asisten riset/vol/milkunC/achaidir/FAO/faostat_area_harvested_IMAGE.xlsx", index=False)
yield_absolute_merge_columns.to_excel("D:/kerja/asisten riset/vol/milkunC/achaidir/FAO/faostat_yield_absolute_IMAGE.xlsx", index=False)
```
