```python
import pandas as pd
import numpy as np
```

```python
INPUT_FILE_PATH = "D:/kerja/asisten riset/vol/milkunC/achaidir/IMAGE PBL/SSP2/natveg2crops_full.xlsx"
df = pd.read_excel(INPUT_FILE_PATH)
df.head()
```

```python
df['type'].unique()
```

### filter

```python
df_filter = df[df['origin'] != 0]
df_filter.head()
```

### groubpy berdasarkan country type

```python
df_groupby = df_filter.groupby(['origin']).sum()
df_groupby.head()
```

### drop columns country

```python
df_index = df_groupby.reset_index()
df_drop = df_index.drop(columns=['country', 'type'])
df_drop.head()
```

```python
df_drop.to_excel("D:/kerja/asisten riset/vol/milkunC/achaidir/IMAGE PBL/SSP2/natveg2crops_origin.xlsx", index=False)
```
