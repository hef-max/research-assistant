```python
# intensity emisi AM2
for index_corps, corps in enumerate(landuse_class_crop):
    #c3ann
    if index_corps == 0:
        area_equ_c3ann += np.multiply.reduce((luh2_static['carea'].to_numpy(),
                                                            np.nan_to_num(luh2_states_worldwide[corps].isel(time=max_year-1).to_numpy())))
        area_equ_all[0][index_corps] += area_equ_c3ann[0]
    #c4ann
    elif index_corps == 1:
        area_equ_c4ann += np.multiply.reduce((luh2_static['carea'].to_numpy(),
                                                            np.nan_to_num(luh2_states_worldwide[corps].isel(time=max_year-1).to_numpy())))
        area_equ_all[0][index_corps] += area_equ_c4ann[0]
    # c3per
    elif index_corps == 2:
        area_equ_c3per += np.multiply.reduce((luh2_static['carea'].to_numpy(),
                                                            np.nan_to_num(luh2_states_worldwide[corps].isel(time=max_year-1).to_numpy())))
        area_equ_all[0][index_corps] += area_equ_c3per[0]
    #c4per
    elif index_corps == 3:
        area_equ_c4per += np.multiply.reduce((luh2_static['carea'].to_numpy(),
                                                            np.nan_to_num(luh2_states_worldwide[corps].isel(time=max_year-1).to_numpy())))
        area_equ_all[0][index_corps] += area_equ_c4per[0]
    #c3nfx
    elif index_corps == 4:
        area_equ_c3nfx += np.multiply.reduce((luh2_static['carea'].to_numpy(),
                                                            np.nan_to_num(luh2_states_worldwide[corps].isel(time=max_year-1).to_numpy())))
        area_equ_all[0][index_corps] += area_equ_c3nfx[0]
    #pastr
    elif index_corps == 5:
        area_equ_pastr += np.multiply.reduce((luh2_static['carea'].to_numpy(),
                                                            np.nan_to_num(luh2_states_worldwide[corps].isel(time=max_year-1).to_numpy())))
        area_equ_all[0][index_corps] += area_equ_pastr[0]
    #range
    elif index_corps == 6:
        area_equ_range += np.multiply.reduce((luh2_static['carea'].to_numpy(),
                                                            np.nan_to_num(luh2_states_worldwide[corps].isel(time=max_year-1).to_numpy())))
        area_equ_all[0][index_corps] += area_equ_range[0]
```
