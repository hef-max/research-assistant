```python
import numpy as np
import pandas as pd
```

```python
fao_stat = pd.read_csv("D:/kerja/asisten riset/vol/milkunC/achaidir/FAO/Production_Crops_Livestock_E_All_Data_analyze.csv")
crop_class = pd.read_excel("D:/kerja/asisten riset/vol/milkunC/achaidir/FAO/Crop Classification_latest.xlsx", engine="openpyxl", skiprows=1)
crop_class = crop_class.drop('Unnamed: 0', axis=1)
crop_class.rename(columns={'FAO Crops': 'Item'}, inplace=True)
```

```python
fao_stat = fao_stat[~fao_stat["Area"].isin(["Americas", "Asia", "Australia and New Zealand", "Africa", "Belgium-Luxembourg", "Central America", "Central Asia", "Caribbean", "Czechoslovakia", "Eastern Africa", "Eastern Asia",
 "Eastern Europe", "European Union (27)","Land Locked Developing Countries", "Least Developed Countries", "Europe", "Low Income Food Deficit Countries", "Melanesia", "Middle Africa", 
  "Net Food Importing Developing Countries", "Northern Africa", "Northern America", "Northern Europe", "Oceania", "Polynesia", "Small Island Developing States", "Serbia and Montenegro",
  "South-eastern Asia", "Southern Africa", "Southern Asia", "Southern Europe", "Sudan (former)", "South America", "USSR", "Western Africa", "Western Asia", "Western Europe", "World", "Yugoslav SFR"])]
```

```python
fao_stat_copy = fao_stat.copy()

prubahan_nama = {
    "China, Hong Kong SAR": "Hong Kong",
    "China, Macao SAR": "Macao", 
    "China, Taiwan Province of": "Taiwan, Province of China",
    "China, mainland": "China",
    "Democratic People's Republic of Korea": "Korea (Democratic People's Republic of)",
    "Democratic Republic of the Congo": "Congo, Democratic Republic of the",
    "Micronesia": "Micronesia (Federated States of)",
    "Micronesia (Federated States of) (Federated States of)": "Micronesia (Federated States of)",
    "Netherlands (Kingdom of the)": "Netherlands",
    "Republic of Korea": "Korea, Republic of",
    "Republic of Moldova": "Moldova, Republic of",
    "Palestine" : "Palestine, State of",
    "Türkiye": "Turkey",
    "United Republic of Tanzania": "Tanzania, United Republic of",
    "Ethiopia PDR" : "Eritrea"
}
keys_list = list(prubahan_nama.keys())
values_list = list(prubahan_nama.values())

# substitute
for i in range(len(keys_list)):
    fao_stat_copy['Area'] = fao_stat_copy['Area'].astype(str).str.replace(keys_list[i], values_list[i])
```

```python
fao_stat_area_harv = fao_stat_copy[fao_stat_copy['Element'] == "Area harvested"]
fao_stat_area_harv = fao_stat_area_harv.fillna(0)
fao_stat_area_harv = fao_stat_area_harv.merge(crop_class, on='Item')
```

```python
fao_stat_yield = fao_stat_copy[fao_stat_copy['Element'] == "Yield"]
fao_stat_yield
# fao_stat_yield = fao_stat_yield.merge(crop_class, on='Item')
# fao_stat_yield['FAO (primary/ processed)'].unique()
```

```python
fao_stat_yield_absolute = fao_stat_yield.copy()
```

```python
fao_stat_yield_absolute = fao_stat_yield.merge(fao_stat_area_harv, on=['Area', 'Item', 'Year'], suffixes=('_yield', '_harv'))
fao_stat_yield_absolute['Value'] = (fao_stat_yield_absolute['Value_yield'] * fao_stat_yield_absolute['Value_harv']) / 1000
```

```python
fao_stat_yield_absolute.to_csv("D:/kerja/asisten riset/vol/milkunC/achaidir/FAO/fao_stat_yield_absolute.csv", index=False)
```

```python
fao_stat_yield_absolute = pd.read_csv("D:/kerja/asisten riset/vol/milkunC/achaidir/FAO/fao_stat_yield_absolute.csv")
fao_stat_yield_absolute
```

```python
fao_stat_yield_absolute['IMAGE Classification_yield'] = fao_stat_yield_absolute['IMAGE Classification_yield'].apply(lambda x: x.replace('and', '&').replace('Plant-based fibers', 'Plant based fibres').replace('Oil, palm fruit', 'Oil & palm fruit').replace('Other non-food, luxury, spices', 'Other non-food & luxury & spices').replace('Vegetables or fruits', 'Vegetables & fruits'))
```

```python
fao_stat_yield_absolute_columns = fao_stat_yield_absolute[['Area', 'Item', 'Year','IMAGE Classification_yield', 'Value']]
fao_stat_yield_absolute_columns = fao_stat_yield_absolute_columns.rename(columns={"IMAGE Classification_yield": "IMAGE Classification"})
fao_stat_yield_absolute_columns = fao_stat_yield_absolute_columns[fao_stat_yield_absolute_columns['Year'] >= 1968]
fao_stat_yield_absolute_columns
# fao_stat_yield_absolute_columns.to_csv("D:/kerja/asisten riset/vol/milkunC/achaidir/FAO/fao_stat_yield_absolute_columns.csv", index=False)
```

```python
fao_stat_yield_absolute_columns_drop = fao_stat_yield_absolute_columns.drop_duplicates(subset=['Area', 'Item', 'Year', 'IMAGE Classification', 'Value'])
fao_stat_yield_absolute_columns_drop
```

```python
# fao_stat_yield_absolute_columns_drop[fao_stat_yield_absolute_columns_drop['IMAGE Classification'] == "Sugar crops"]
```

```python
step1_copy = fao_stat_yield_absolute_columns_drop.copy()
step1_copy['Value'] = step1_copy['Value'].fillna(0)
step1_copy['Value'] = 0
step1_copy['Value'] = fao_stat_yield_absolute_columns_drop.groupby(['Area', "Year", 'IMAGE Classification'])['Value'].transform('sum')
step1_copy
```

```python
step1_copy_drop = step1_copy.drop_duplicates(subset=['Area', 'Year', 'IMAGE Classification', 'Value'])
step1_copy_drop
```

```python
# step1_copy_drop[step1_copy_drop['IMAGE Classification'] == "Sugar crops"]
```

```python
step1_copy_columns = step1_copy_drop[['Area', 'Year', 'IMAGE Classification', 'Value']]
step2_copy = step1_copy_columns.copy()
eight_year_ranges = [(y, y+4) for y in range(1968, 2022, 5)]

for year in eight_year_ranges:
    mask = (step1_copy_columns['Year'] >= year[0]) & (step1_copy_columns['Year'] <= year[1]) #5years
    step2_copy.loc[mask, 'Value'] = step1_copy_columns[mask].groupby(['Area', 'IMAGE Classification'])['Value'].transform('mean')
step2_copy
```
