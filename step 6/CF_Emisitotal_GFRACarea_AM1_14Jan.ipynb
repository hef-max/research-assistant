```python
columns = [x for x in range(1975, 2021, 5)]
data1['type'] = data1['type'].replace("palm oil", "oil & palm fruit")
data1 = data1[~data1['type'].str.contains("pasture")]
data1['type'] = data1['type'].str.lower()

# data1['IMAGE Classification Region'] = data1['IMAGE Region Name'] + "-" + data1['type']
data1_group = data1.groupby(['IMAGE Region Name','type']).sum()
data1_group = data1_group.reset_index()
data1_group_selected = data1_group[['IMAGE Region Name', 'type', 1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015, 2020]]
emissions_melt = data1_group_selected.melt(id_vars=['IMAGE Region Name', 'type'], var_name='time', value_name='value')
emissions_melt['value'] = emissions_melt['value']*1e9
emissions_melt
```

```python
data2_melted = pd.melt(data2, id_vars=['country', 'type'], var_name='time', value_name='value')
data2_melted
```

```python
ccode_iso = pd.read_excel("Z:/achaidir/Country Grids/ISO-3166-Country-Code_Final.xlsx")
data2_melted_region = pd.merge(left=ccode_iso, right=data2_melted, how="left", left_on="ISO Country", right_on="country")
data2_melted_region
```

```python
data2_melted_region_columns = data2_melted_region[['country', 'IMAGE Region Name', 'type', 'time', 'value']]
data2_melted_region_columns_groupby = data2_melted_region_columns.groupby(['IMAGE Region Name','type', 'time']).sum()
data2_melted_region_index = data2_melted_region_columns_groupby.reset_index()
data2_melted_region_index['type'] = data2_melted_region_index['type'].str.lower()
data2_melted_region_index
```

```python
CF_emisi_total_GFRAC_area = emissions_melt.copy()
CF_emisi_total_GFRAC_area = CF_emisi_total_GFRAC_area.drop(columns=['value'])
IMAGERegionName = CF_emisi_total_GFRAC_area["IMAGE Region Name"].to_list()
time = CF_emisi_total_GFRAC_area["time"].to_list()
type_ngfbfc = CF_emisi_total_GFRAC_area["type"].to_list()

for region, time, ngfbfc in zip(IMAGERegionName, time, type_ngfbfc):
    try:
        data1 = emissions_melt[(emissions_melt['IMAGE Region Name'] == region) & (emissions_melt['time'] == time) & (emissions_melt['type'] == ngfbfc)]['value'].to_numpy()
        data2 = data2_melted_region_index[(data2_melted_region_index['IMAGE Region Name'] == region) & (data2_melted_region_index['time'] == time) & (data2_melted_region_index['type'] == ngfbfc)]['value'].to_numpy()
        if data1.size == 0:
            data1 = [0]
        if data2.size == 0: 
            data2 = [0]
        mask = emissions_melt[(emissions_melt['IMAGE Region Name'] == region) & (emissions_melt['time'] == time) & (emissions_melt['type'] == ngfbfc)].index
        CF_emisi_total_GFRAC_area.loc[mask, 'CF'] = data1[0] / data2[0]
    except KeyError as e:
        print(f"Error: {e} for region {region}")
```
