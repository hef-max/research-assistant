```python
import pandas as pd

grapc_sheet_names = [" Wheat"," Rice"," Maize"," Tropical cereals",
                     " Other temperate cereals"," Pulses"," Soybeans",
                     " Temperate oil crops"," Tropical oil crops",
                     " Temperate roots & tubers"," Tropical roots & tubers",
                     " Sugar crops"," Oil & palm fruit"," Vegetables & fruits",
                     " Othernon-food&luxury spices"," Plant based fibres"]

INPUT_FILE_PATH = "X:/cadlan/stream_2/Step5/GRAPC_area_full.xlsx"
OUTPUT_FILE_PATH = "X:/cadlan/stream_2/Step5/GRAPC_area_full_concat.xlsx"

dataframes = []
for sheet_name in grapc_sheet_names:
    df = pd.read_excel(INPUT_FILE_PATH, sheet_name=sheet_name)
    dataframes.append(df)

tahun = list(range(1975, 1990, 5))
tahun.append('origin')
tahun.append('country')

merged_df = pd.concat(dataframes, axis=0, ignore_index=True).fillna(0)
merged_df.to_excel(OUTPUT_FILE_PATH, index=False)
```

```python
country_code = pd.read_excel("Z:/achaidir/Country Grids/ISO-3166-Country-Code_Final.xlsx")
```

```python
df2 = pd.read_excel("X:/cadlan/stream_2/Step5/GRAPC_area_full_concat.xlsx")
```

```python
df2['type'].unique()
```

```python
#rubah/ join country ke region

df2['NGFBFC'] = df2['type'].apply(lambda x: x.replace('and', '&').replace('Plant-based fibers', 'Plant based fibres').replace('Oil, palm fruit', 'Oil & palm fruit').replace('Othernon-food&luxury spices', 'Other non-food & luxury & spices').replace('Vegetables or fruits', 'Vegetables & fruits'))
df2_merge = pd.merge(left=df2, right=country_code, left_on="country", right_on="ISO Country")
```

```python
df2_merge['ISO Country'].unique()
```

```python
#melt kebawah sebagai row 

time = list(range(1970, 2025, 5)) 
df2_pivot = pd.pivot_table(data=df2_merge, values=time, index=['IMAGE Region Name', 'type'], fill_value=0, aggfunc='sum')
df2_index = df2_pivot.reset_index()
df2_melt = df2_index.melt(id_vars=['IMAGE Region Name', 'type'], var_name='time', value_name='value')
df2_melt.head()
```

```python
df2_melt['IMAGE Region Name'].unique()
```

```python
df2_melt_groupby = df2_melt.groupby(["IMAGE Region Name", "type", "time"]).sum()
df2_melt_groupby_index = df2_melt_groupby.reset_index()
df2_melt_groupby_index
```

```python
df2_melt_groupby_index['IMAGE Region Name'].unique()
```

```python
df2_melt_groupby_index = df2_melt_groupby_index.rename(columns={"type" : "NGFBFC"})
df2_melt_groupby_index
```

```python
df2_melt_groupby_index.to_csv("X:/cadlan/stream_2/Step5/GRAPC_production_region_melted.csv", index=False)
```
