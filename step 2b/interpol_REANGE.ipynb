```python
import xarray
import numpy as np
import pandas as pd
```

```python
path_LUH2 = 'D:/kerja/asisten riset/vol/milkunC/achaidir/LUH2 2022/states.nc'
luh2_states = xarray.open_dataset(path_LUH2, engine="netcdf4", decode_times=False)

luh2_states_worldwide = luh2_states.isel(time=slice(1000, 1173))
luh2_states_worldwide['time'] = pd.date_range(start="1850-01-01", end="2022-01-01", freq='YS')

luh2_added_states_worldwide = xarray.open_dataset("D:/kerja/asisten riset/vol/milkunC/achaidir/LUH2 2022/multiple-states_input4MIPs_landState_ScenarioMIP_UofMD-IMAGE-ssp119-2-1-f_gn_2015-2100.nc",\
                                        engine="netcdf4", decode_times=False)

luh2_added_states_worldwide = luh2_added_states_worldwide.drop_vars(['lat_bounds', 'lon_bounds', 'time_bnds'])

luh2_added_states_worldwide.coords['time'] = pd.date_range(start='2015-01-01', end='2100-01-01', freq='YS')
```

```python
rangeland = xarray.open_dataset("D:/kerja/asisten riset/vol/milkunarc/cadlan/Analysis2a_sum/RANGELAND-1970-2100.nc")
nbp = ['branches', 'stems', 'leaves', 'roots', 'litter', 'humus', 'charcoal', 'short lifetime timber', 'long lifetime timber']
```

```python
def carstock_interp(carstock_class, lat_idx):
    '''
    Function to interpolate carbon stocks between two given time.
    Carbon class in parameter is:
    - leaves.
    - branches.
    - stems.
    - roots.
    
    Parameter:
    carstock_class: carbon stocks class that will be interpolated.
    
    Return:
    An array with size (131, 1, 1, 38), where the meaning of the size is: 131 stand for 131 years (1970 - 2100), 1 after 6 stand for one latitude each year, the second 1 stand for 1
    longitude each latitude and each year, and 38 stand for there are 38 elements inside the longitude array.
    '''
    # Define placeholder array with NaN value in the element
    arr = np.full((131, 1, 1, 1440), fill_value=np.nan)
    
    # Define carbon stocks variable given year
    range_any_1970 = np.array(([rangeland.sel(NBP=carstock_class).isel(lat=lat_idx, time=0)['extensive grassland'].to_numpy()]))
    range_any_1975 = np.array(([rangeland.sel(NBP=carstock_class).isel(lat=lat_idx, time=1)['extensive grassland'].to_numpy()]))
    range_any_1980 = np.array(([rangeland.sel(NBP=carstock_class).isel(lat=lat_idx, time=2)['extensive grassland'].to_numpy()]))
    range_any_1985 = np.array(([rangeland.sel(NBP=carstock_class).isel(lat=lat_idx, time=3)['extensive grassland'].to_numpy()]))
    range_any_1990 = np.array(([rangeland.sel(NBP=carstock_class).isel(lat=lat_idx, time=4)['extensive grassland'].to_numpy()]))
    range_any_1995 = np.array(([rangeland.sel(NBP=carstock_class).isel(lat=lat_idx, time=5)['extensive grassland'].to_numpy()]))
    range_any_2000 = np.array(([rangeland.sel(NBP=carstock_class).isel(lat=lat_idx, time=6)['extensive grassland'].to_numpy()]))
    range_any_2005 = np.array(([rangeland.sel(NBP=carstock_class).isel(lat=lat_idx, time=7)['extensive grassland'].to_numpy()]))
    range_any_2010 = np.array(([rangeland.sel(NBP=carstock_class).isel(lat=lat_idx, time=8)['extensive grassland'].to_numpy()]))
    range_any_2015 = np.array(([rangeland.sel(NBP=carstock_class).isel(lat=lat_idx, time=9)['extensive grassland'].to_numpy()]))
    range_any_2020 = np.array(([rangeland.sel(NBP=carstock_class).isel(lat=lat_idx, time=10)['extensive grassland'].to_numpy()]))
    range_any_2025 = np.array(([rangeland.sel(NBP=carstock_class).isel(lat=lat_idx, time=11)['extensive grassland'].to_numpy()]))
    range_any_2030 = np.array(([rangeland.sel(NBP=carstock_class).isel(lat=lat_idx, time=12)['extensive grassland'].to_numpy()]))
    range_any_2035 = np.array(([rangeland.sel(NBP=carstock_class).isel(lat=lat_idx, time=13)['extensive grassland'].to_numpy()]))
    range_any_2040 = np.array(([rangeland.sel(NBP=carstock_class).isel(lat=lat_idx, time=14)['extensive grassland'].to_numpy()]))
    range_any_2045 = np.array(([rangeland.sel(NBP=carstock_class).isel(lat=lat_idx, time=15)['extensive grassland'].to_numpy()]))
    range_any_2050 = np.array(([rangeland.sel(NBP=carstock_class).isel(lat=lat_idx, time=16)['extensive grassland'].to_numpy()]))
    range_any_2055 = np.array(([rangeland.sel(NBP=carstock_class).isel(lat=lat_idx, time=17)['extensive grassland'].to_numpy()]))
    range_any_2060 = np.array(([rangeland.sel(NBP=carstock_class).isel(lat=lat_idx, time=18)['extensive grassland'].to_numpy()]))
    range_any_2065 = np.array(([rangeland.sel(NBP=carstock_class).isel(lat=lat_idx, time=19)['extensive grassland'].to_numpy()]))
    range_any_2070 = np.array(([rangeland.sel(NBP=carstock_class).isel(lat=lat_idx, time=20)['extensive grassland'].to_numpy()]))
    range_any_2075 = np.array(([rangeland.sel(NBP=carstock_class).isel(lat=lat_idx, time=21)['extensive grassland'].to_numpy()]))
    range_any_2080 = np.array(([rangeland.sel(NBP=carstock_class).isel(lat=lat_idx, time=22)['extensive grassland'].to_numpy()]))
    range_any_2085 = np.array(([rangeland.sel(NBP=carstock_class).isel(lat=lat_idx, time=23)['extensive grassland'].to_numpy()]))
    range_any_2090 = np.array(([rangeland.sel(NBP=carstock_class).isel(lat=lat_idx, time=24)['extensive grassland'].to_numpy()]))
    range_any_2095 = np.array(([rangeland.sel(NBP=carstock_class).isel(lat=lat_idx, time=25)['extensive grassland'].to_numpy()]))
    range_any_2100 = np.array(([rangeland.sel(NBP=carstock_class).isel(lat=lat_idx, time=26)['extensive grassland'].to_numpy()]))
    
    # Change the value of first and last placeholder array with defined carbon stocks variable
    arr[0] = range_any_1970
    arr[5] = range_any_1975
    arr[10] = range_any_1980
    arr[15] = range_any_1985
    arr[20] = range_any_1990
    arr[25] = range_any_1995
    arr[30] = range_any_2000
    arr[35] = range_any_2005
    arr[40] = range_any_2010
    arr[45] = range_any_2015
    arr[50] = range_any_2020
    arr[55] = range_any_2025
    arr[60] = range_any_2030
    arr[65] = range_any_2035
    arr[70] = range_any_2040
    arr[75] = range_any_2045
    arr[80] = range_any_2050
    arr[85] = range_any_2055
    arr[90] = range_any_2060
    arr[95] = range_any_2065
    arr[100] = range_any_2070
    arr[105] = range_any_2075
    arr[110] = range_any_2080
    arr[115] = range_any_2085
    arr[120] = range_any_2090
    arr[125] = range_any_2095
    arr[-1] = range_any_2100
    
    # Save the array value into new dataframe
    years = list(np.linspace(1970, 2100, 131, dtype="int64"))
    car_stock_any_df = pd.DataFrame()
    for i, year in enumerate(years):
        car_stock_any_df.insert(i, f'{year}', arr[i][0][0], True)
    
    # Interpolate the data and save it into new interploate variable
    interp_boreal_forest_any = car_stock_any_df.interpolate(method="linear", axis=1)
    
    # Tranlate the interploate data into new array
    arr_interp_carstock_any = []
    for year in years:
        arr_interp_carstock_any.append([[interp_boreal_forest_any[f'{year}']]])
    
    return np.array(arr_interp_carstock_any)
```

```python
def convert_interp2arr(carstock_class):
    '''
    Convert from interpolation per latitude in 131 year into the correct form (131 year, 51 latitude, 38 element of longitude length)
    Parameter:
    carstock_class: carbon stock class
    
    Return:
    An array with size (131, 51, 38)
    '''
    interp_any = [] # -> (51, 131, 1, 1, 38)
    for i in range(720):
        interp_any.append(carstock_interp(f'{carstock_class}', i))
 #720ganti1440   
    arr_interp_any = np.array(interp_any)
    carstock_interp_any_year = [] #-> (131, 51, 1, 1, 1, 38)
    for year in range(131):
        carstock_interp_any_year.append([])
        for i in range(len(arr_interp_any)):
            carstock_interp_any_year[year].append(arr_interp_any[i][year])
    carstock_interp_any_year = np.array(carstock_interp_any_year).reshape(131, 720, 1440)
    
    return carstock_interp_any_year
```

```python
range_interp_branches = convert_interp2arr('branches')
range_interp_leaves = convert_interp2arr('leaves')
range_interp_stems = convert_interp2arr('stems')
range_interp_roots = convert_interp2arr('roots')
range_interp_charcoal = convert_interp2arr('charcoal')
range_interp_litter = convert_interp2arr('litter')
range_interp_humus = convert_interp2arr('humus')
range_interp_short_lifetime_timber = convert_interp2arr('short lifetime timber')
range_interp_long_lifetime_timber = convert_interp2arr('long lifetime timber')

range_interp_branches_expand = np.expand_dims(range_interp_branches, axis=-1)
range_interp_leaves_expand = np.expand_dims(range_interp_leaves, axis=-1)
range_interp_stems_expand = np.expand_dims(range_interp_stems, axis=-1)
range_interp_roots_expand = np.expand_dims(range_interp_roots, axis=-1)
range_interp_charcoal_expand = np.expand_dims(range_interp_charcoal, axis=-1)
range_interp_litter_expand = np.expand_dims(range_interp_litter, axis=-1)
range_interp_humus_expand = np.expand_dims(range_interp_humus, axis=-1)
range_interp_short_lifetime_timber_expand = np.expand_dims(range_interp_short_lifetime_timber, axis=-1)
range_interp_long_lifetime_timber_expand = np.expand_dims(range_interp_long_lifetime_timber, axis=-1)

range_interp_concat = np.concatenate([range_interp_branches_expand, range_interp_leaves_expand, range_interp_stems_expand, range_interp_roots_expand,
                                      range_interp_charcoal_expand, range_interp_litter_expand, range_interp_humus_expand, range_interp_short_lifetime_timber_expand,
                                      range_interp_long_lifetime_timber_expand], axis=-1)
```

```python
carbon_stock_interp = xarray.Dataset({
        "range_interp":(["time", "lat", "lon", "NBP"], range_interp_concat)
    },
    coords={
        "lon":luh2_states_worldwide.coords["lon"].to_numpy(),
        "lat":luh2_states_worldwide.coords["lat"].to_numpy(),
        "time":pd.date_range("1970-01-01", "2100-01-01", freq='YS'),
        "NBP": nbp
    })

# %%
carbon_stock_interp.to_netcdf("D:/kerja/asisten riset/vol/milkunarc/cadlan/Analysis2b_interpolate/CARBON-STOCK-RANGE-INTERPOLATION-WORLDWIDE-1970_2100.nc", mode='w', format="NETCDF4")
```
